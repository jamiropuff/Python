# Ejercicio 1

Generar una función y utilizar esta para recodificar la variable “país de origen” (1, 2, 3) en el dataframe a través de la función apply.

```python
# Importamos Numpy y Pandas para crear un DataFrame
import numpy as np
import pandas as pd

# Importamos los datos del CSV y los mostramos en un DataFrame
df = pd.read_csv('/content/autos.csv', sep=';')
df
```

```python
# Generamos una función de tipo lambda para recodificar la variable “país de origen” (1, 2, 3) en el dataframe a través de la función apply.
df['pais_origen_r'] = df['Pais_Origen'].apply(lambda x:
                                                    1 if x=="USA" else (
                                                        2 if x=="Japan" else (
                                                            3 if x=="Europe" else 0
                                                        )
                                                    )
)
df
```

```python
#3. Si var 1 > var 2 → 5, caso contrario 3
# Creando una Función para verificar si se var 1 es mayor a var 2
def varMayor(var1,var2):
  '''
  Devuelve 5 si la variable 1 es mayor a la variable 2, en caso contrario devuelve 3

  Parámetros
  ----------
  var1
  en *var* puedes ingresar un número de tipo 'int' o 'float'.

  var2
  en *var2* puedes ingresar un número de tipo 'int' o 'float'.

  Ejemplos:
  ----------
  >> varMayor(var1=15,var=2)
  >> 5

  >> varMayor(var1=8,var2=25)
  >> 3
  '''

  variable_mayor = ""

  if var1 > var2:
    variable_mayor = 5
  else:
    variable_mayor = 3

  return variable_mayor

# Convertimos los Cilindros en una Lista
cilindros = list(autos_num.Cilindros)

# Convertimos la Aceleracion en una Lista
aceleracion = list(autos_num.Aceleracion)

total_registros = len(cilindros)

# Verificamos si la variable 1 es mayor a la variable 2
col3 = []
for i in range(0,total_registros):
  #print("i: ",i)
  var1 = cilindros[i]
  var2 = aceleracion[i]
  var_mayor = varMayor(var1,var2)
  col3.append(var_mayor)

# Agregamos una columna con la Meta Superada
#len(col3)
autos_num_variables = (autos_num_variables.assign( c3 = col3 ))
autos_num_variables
```

# Ejercicio 3

Crear una función que permita generar 3 variables cualitativas
ordinales a partir de las variables numéricas con los siguientes criterios:
★ Si es mayor a la “media + 1*desviación estándar” → Alto
★ Si está entre “media - 1*desviación estándar” y “media +
1*desviación estándar” → Medio
★ Si es menor a la “media - 1*desviación estándar” → Bajo
Nota: Mostrar con las nuevas variables generadas.
Elegir con libertad las variables a trabajar.
La función debe contar con su respectiva documentación.

```python
# Creando una Función para generar 3 variables cualitativas ordinales a partir de variables numéricas
def varCualitativasOrdinales(variable):
  '''
  Devuelve 3 variables cualitativas a partir de una variable numérica:

  - Variable 1: Alto -> Si es mayor a la “media + 1*desviación estándar”
  - Variable 2: Medio -> Si está entre “media - 1*desviación estándar” y “media + 1*desviación estándar”
  - Variable 3: Bajo -> Si es menor a la “media - 1*desviación estándar”

  Parámetros
  ----------
  variable
  en *variable* puedes ingresar una columna de un dataFrame df['Cilindros'].

  Ejemplos:
  ----------
  >> varCualitativasOrdinales(df['Cilindros'])
  >> Medio
  >> Bajo
  >> Medio
  >> Alto
  >> Alto
  >> Medio

  Nos regresa el comparativo de toda la columna y podemos agregarlo posteriormente a nuestro dataFrame en caso de ser necesario
  '''

  # Realizamos los calculos necesarios
  calculo1 = lambda x: ( x.mean() + (1 * x.std()) )
  calculo2 = lambda x: ( x.mean() - (1 * x.std()) )

  # Asignamos los calculos a variables
  calc1 = calculo1(variable)
  calc2 = calculo2(variable)

  # Declaramos las variables que vamos a usar
  col = []
  var_cualitativa = ''
  contador = 0

  # Recorremos el listado y comparamos para obtener los valores Alto, Medio y Bajo
  for i in variable:

    if(i > calc1):
      var_cualitativa = 'Alto'

    if( i > calc2 ):
      contador = 1

    if( i < calc1 ):
      contador = contador + 1

    if(contador == 2):
        var_cualitativa = 'Medio'
        contador = 0

    if(i < calc2):
      var_cualitativa = 'Bajo'

    # Almacenamos los valores en una lista
    col.append(var_cualitativa)

  # Regresamos el valor
  return col

# Data numerica
autos_cualitativa = df.loc[:, ['Millas_Por_Galon', 'Cilindros', 'Pais_Origen', 'Peso_En_Libras', 'Desplazamiento', 'Aceleracion','Caballo_De_Fuerza','pais_origen_r']]
autos_cualitativa.head(15)

# Agregamos las columnas necesarias con los calculos realizados
autos_cualitativa['millas_por_galon_cualitativa'] = varCualitativasOrdinales(autos_cualitativa['Millas_Por_Galon'])
autos_cualitativa['cilindros_cualitativa'] = varCualitativasOrdinales(autos_cualitativa['Cilindros'])
autos_cualitativa['peso_en_libras_cualitativo'] = varCualitativasOrdinales(autos_cualitativa['Peso_En_Libras'])
autos_cualitativa['desplazamiento_cualitativo'] = varCualitativasOrdinales(autos_cualitativa['Desplazamiento'])
autos_cualitativa['aceleracion_cualitativo'] = varCualitativasOrdinales(autos_cualitativa['Aceleracion'])
autos_cualitativa['caballos_de_fuerza_cualitativo'] = varCualitativasOrdinales(autos_cualitativa['Caballo_De_Fuerza'])
autos_cualitativa
```

# Ejercicio 5

Crear una función en Python que devuelva un mapa de calor con la
correlación de todas las variables numéricas (incluye las generadas),
según país.

```python
def graficoMapaCalor(columnas: list, data:pd.DataFrame): # parámetros
  '''
  Gráfico de Mapa de Calor con la correlación de todas las variables numéricas.

  Parámetros
  ----------

  `columnas` : lista con los nombres de las variables numéricas.
  `data` : de tipo DataFrame.

  Esta función graficará un  gráfico de Mapa de Calor con la correlación de todas las variables numéricas.

  Ejemplos:
  --------

  '''

  correlacion = data.corr()

  mask = np.triu(np.ones_like(correlacion, dtype=bool ))

  sns.heatmap(correlacion,
            mask = mask,
            xticklabels = correlacion.columns.values,
            yticklabels = correlacion.columns.values,
            vmax=1,
            center = 0,
            square = True,
            linewidths=.1,
            annot = True,
            cmap = "Blues"
            )

```
