# Ejercicio 1

```python
# Importamos Numpy y Pandas para crear un DataFrame
import numpy as np
import pandas as pd

# Creo un DataFrame con tres columnas, Producto, Cantidad, Precio Unitario
data = {
    'Producto':["CALIFORNIA DREAMS", "CALIFORNIA PLAYA", "NEW YORK", "NUEVO MÉXICO", "NAVY", "AIR FORCE"],
    'Cantidad':[2, 1, 4, 3, 5, 2],
    'Precio_Unitario':[286, 405, 260, 234, 180, 220]
    }

# Creamos el DataFrame con las fechas de venta
df = pd.DataFrame(data, index=["2024-03-22","2024-03-21","2024-03-20","2024-03-19","2024-03-18","2024-03-17"])
df
```

```python
# Obtenemos el Total a pagar
# Agregamos una columna al final con el total
df = (
    df.assign(Total = lambda x: x.Cantidad * x.Precio_Unitario)
)
df
```

```python
# Calculamos el Precio con Descuento del 5%
# Agregamos una columna al final con el precio con descuento
df = (
    df.assign(Precio_Descuento = lambda x: x.Total - (x.Total * .05))
)
df
```

# Ejercicio 2

```python
# Importamos Numpy y Pandas para crear un DataFrame
import numpy as np
import pandas as pd

# Creo un DataFrame con tres columnas, Producto, Cantidad, Precio Unitario
data = {
    'Ingreso':[30000, 25000, 75000, 150000, 20000],
    }

# Creamos el DataFrame con las clientes
df2 = pd.DataFrame(data, index=["DUELAZO","ADVAITA","GRUPO_BIMBO","UNILEVER","AVANA"])
df2
```

```python
# Obtenemos el mínimo ingreso, máximo ingreso y la media del ingreso
minimo = df2.Ingreso.min()
maximo = df2.Ingreso.max()
media = df2.Ingreso.mean()

# Obtenemos a los clientes de cada caso mínimo y máximo
cliente_minimo = df2.index[(df2.Ingreso == minimo)].tolist()
cliente_maximo = df2.index[(df2.Ingreso == maximo)].tolist()

# Mostramos los valores
print("Mínimo: ",cliente_minimo[0]," ",minimo)
print("Máximo: ",cliente_maximo[0]," ",maximo)
print("Media: ",media)
```

# Ejercicio 3

```python
# Importamos Numpy y Pandas para crear un DataFrame
import numpy as np
import pandas as pd

# Creo un DataFrame con tres columnas, Mes, Ventas, Gastos
data = {
    'Mes':["Enero", "Febrero", "Marzo", "Abril", "Mayo"],
    'Ventas':[30500, 35600, 28300, 33900, 42500],
    'Gastos':[22000, 23450, 18100, 35700, 32450]
    }

# Creamos el DataFrame con las fechas de venta
df3 = pd.DataFrame(data)
df3
```

```python
# 1. Columna 1 = Gasto/Ventas*100.

df3 = ( df3.assign( Columna1 = lambda x: (x.Gastos / x.Ventas) * 100 ) )
df3
```

```python
# 2. Columna 2 = Si Ventas>= 30000, imprimir “Meta Superada”, caso contrario imprimir “Meta no Superada”.

# Creando una Función para verificar si se superaron las metas
def metaSuperada(venta,meta):
  '''
  Devuelve el texto Meta Superada o Meta no Superada, dependiendo del valor de la Venta y el valor de la Meta

  Parámetros
  ----------
  venta
  en *venta* puedes ingresar un número de tipo 'int' o 'float'.

  meta
  en *meta* puedes ingresar un número de tipo 'int' o 'float'.

  Ejemplos:
  ----------
  >> metaSuperada(venta=40000,meta=30000)
  >> Meta Superada

  >> metaSuperada(venta=23000,meta=30000)
  >> Meta no Superada
  '''

  meta_superada = ""

  if venta >= meta:
    meta_superada = "Meta Superada"
  else:
    meta_superada = "Meta no Superada"

  return meta_superada

# Convertimos las Ventas en una Lista
ventas = list(df3.Ventas)

# Verificamos si las ventas superaron la meta
col2 = []
for i in ventas:
  meta = metaSuperada(i,30000)
  col2.append(meta)

# Agregamos una columna con la Meta Superada
df3 = (df3.assign( Columna2 = col2 ))
df3
```

```python
# Columna 3 = Si Gasto>Ventas, calcular el sobre gasto.

# Creando una Función para calcular el sobregasto
def sobreGasto(gasto,venta):
  '''
  Devuelve el sobregasto, dependiendo de si el Gasto es mayor a la Venta

  Parámetros
  ----------
  gasto
  en *gasto* puedes ingresar un número de tipo 'int' o 'float'.

  venta
  en *venta* puedes ingresar un número de tipo 'int' o 'float'.

  Ejemplos:
  ----------
  >> sobreGasto(gasto=35000,venta=25000)
  >> 10000
  '''

  sobregasto = 0

  if gasto > venta:
    sobregasto = (gasto - venta)

  return sobregasto


# Convertimos los Gastos y las Ventas en una Lista
gastos = list(df3.Gastos)
ventas = list(df3.Ventas)

# Verificamos si existe un sobregasto
col3 = []
for i in gastos:

  indice = gastos.index(i)
  gasto = i
  venta = ventas[indice]

  sobre_gasto = sobreGasto(gasto,venta)
  col3.append(sobre_gasto)

# Agregamos una columna con la Meta Superada
df3 = (df3.assign( Columna3 = col3 ))
df3
```

```python
# 4. Obtener el total de las ventas de los últimos 3 meses
ventas = list(df3[-3:]["Ventas"])
#print(ventas)

# Sumamos las ventas
sum = 0
for i in ventas:
  sum += i

# Imprimimos en pantalla la Suma Total de las Ventas de los últimos 3 meses
print("Total de las Ventas de los últimos 3 meses: ",sum)
print("")

# Imprimimos en pantalla el total de las Ventas de los últimos 3 meses
df3_ventas = df3[-3:][["Mes","Ventas"]]
df3_ventas

```

```python
# 5. Obtener el total de gastos de febrero a abril
gastos = list(df3[:]["Gastos"])
#print(gastos)

# Sumamos los gastos
sum = 0
for i in gastos:
  sum += i

# Imprimimos en pantalla la Suma Total de las Ventas de los últimos 3 meses
print("Total de los Gastos de febrero a abril: ",sum)
print("")

# Imprimimos en pantalla el total de las Ventas de los últimos 3 meses
df3_gastos = df3[1:4][["Mes","Gastos"]]
df3_gastos
```
